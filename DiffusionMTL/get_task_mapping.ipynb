```python
import numpy as np
import torch
```

```python
onelabel = np.load('onelabel.npy',allow_pickle=True).item()
randomlabels = np.load('randomlabels.npy',allow_pickle=True).item()
```

```python
onelabel.keys()
```

```python
tasks = ['semseg', 'human_parts', 'sal', 'normals', 'edge']

new_onelabel = {}
for t_idx, task in enumerate(tasks):
    for it in onelabel[task]:
        if it not in new_onelabel.keys():
            new_onelabel[it] = [0 for _ in range(len(tasks))]
            new_onelabel[it][t_idx] = 1
        else:
            new_onelabel[it][t_idx] = 1
            
new_randomlabels = {}
for t_idx, task in enumerate(tasks):
    for it in randomlabels[task]:
        if it not in new_randomlabels.keys():
            new_randomlabels[it] = [0 for _ in range(len(tasks))]
            new_randomlabels[it][t_idx] = 1
        else:
            new_randomlabels[it][t_idx] = 1
```

```python
torch.save(new_randomlabels, 'new_randomlabels.pth')
torch.save(new_onelabel, 'new_onelabel.pth')
```

```python
saved_new_randomlabels = torch.load('new_randomlabels.pth')
saved_new_onelabel = torch.load('new_onelabel.pth')
```
